```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate delta-v
def delta_v(ve, m0, mf):
    return ve * np.log(m0 / mf)

# Parameters (you can change these values to experiment)
ve = 3000  # effective exhaust velocity in m/s
m0 = 500000  # initial mass in kg
mf_values = np.linspace(100000, m0, 500)  # final mass values in kg

# Calculate delta-v for each final mass
delta_v_values = delta_v(ve, m0, mf_values)

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(mf_values, delta_v_values)
plt.title("Tsiolkovsky's Rocket Equation")
plt.xlabel("Final Mass (kg)")
plt.ylabel("Delta-v (m/s)")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate delta-v
def delta_v(ve, m0, mf):
    return ve * np.log(m0 / mf)

# Parameters (you can change these values to experiment)
ve = 3000  # effective exhaust velocity in m/s
m0 = 500000  # initial mass in kg (including propellant)
mf = 100000  # final mass in kg (after fuel is burned)
burn_time = 200  # total burn time in seconds
time_steps = 500  # number of time steps to simulate

# Time array
time = np.linspace(0, burn_time, time_steps)

# Calculate the mass at each time step
mass_flow_rate = (m0 - mf) / burn_time  # rate at which mass is lost (kg/s)
mass = m0 - mass_flow_rate * time  # mass at each time step

# Calculate the velocity at each time step
velocity = ve * np.log(m0 / mass)

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(time, velocity)
plt.title("Rocket Velocity Over Time")
plt.xlabel("Time (s)")
plt.ylabel("Velocity (m/s)")
plt.grid(True)
plt.show()
```
